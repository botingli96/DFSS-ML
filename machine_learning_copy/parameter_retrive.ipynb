```python
import tensorflow as tf
from tensorflow.keras.models import load_model

# Define custom loss function again as it was used in the model
def custom_loss(y_true, y_pred):
    B_true, E_true = y_true[:, :1], y_true[:, 1:]
    B_pred, E_pred = y_pred[:, :1], y_pred[:, 1:]
    loss_B = tf.reduce_mean(tf.square(B_true - B_pred), axis=0)
    loss_E = tf.reduce_mean(tf.square(E_true - E_pred), axis=0)
    weight_B = tf.constant([1.0])
    weight_E = tf.constant([1.0, 1.0, 1.0])
    total_loss = tf.reduce_sum(weight_B * loss_B) + tf.reduce_sum(weight_E * loss_E)
    return total_loss

# Load the model
model_save_path = "/home/botingl/machine learning/FCNN_Bz_run3_model.h5"
model = load_model(model_save_path, custom_objects={'custom_loss': custom_loss})

# Retrieve the model architecture
model.summary()

# Retrieve optimizer information
optimizer_config = model.optimizer.get_config()
learning_rate = optimizer_config['learning_rate']

# Print optimizer information and learning rate
print(f"Optimizer: {optimizer_config['name']}")
print(f"Learning Rate: {learning_rate}")

# Retrieve batch size if needed (not directly stored in the .h5)
# You would need to track the batch size manually during training or store it separately.
```
